
# Course Overview and Schedule

Introduction to Python (Unit 2) Fundamentals is a beginners course, intend to be a 4 week course taking 12 - 16 hours to complete and requires only access to a computer with a modern web browser and an internet connection. The course is not intended for the experienced programmer. The student will be introduced to Data Structures and Files in Python 3 and be ready for more advanced Python learning upon completion.

**Students**
**Should complete Introduction to Python: (Unit 1) Absolute Beginner as prerequisite** Introduction to Python: (Unit 2) Fundamentals is designed to be accessible to beginners with basic Python coding experience creating simple console functions that use input, print output, loops and math operators.

**System Requirements**
All learning activities can be done online. All that is needed is a computer with an internet connection and a modern web browser.

**Schedule**
Introduction to Python: (Unit 2) Fundamentals is a self-paced course that consists of four instructional modules (1-4). All four of the modules are available when the course begins, and you can work your way through the modules at your own pace (the course will close after about 6 months).

While each student will complete this course at their own pace, we anticipate that an average beginner student will be able to complete the course instruction and core lab assignments in about 12 - 16 hours (averaging 3-4 hours per module). We recommend that you attempt to complete one module per week, and rather than pushing on to the next module, that you allocate any additional study time that you have (for a particular week) to exploring suggested resources and experimenting with what you've learned during the labs. A slower or faster pace may work well for some students, but we believe that this approach will be affective for most students. Module 5 is a Final coding Project.

**Module Structure**
**Modules 1 - 4**

Modules 1-4 contain 4 sections of Instructional Content each, with Concept Videos, Sample Code, and Practice Tasks. At the end of the Instructional Content sections are T/F and multiple choice self-check questions (these account for 50% of the grade).

The last section of each Module contains Practice Content and a Required Module level Coding Assignment. The Practice and Required Coding Assignment have corresponding Jupyter Notebook versions (containing the same content) where you can run examples and complete Tasks. For the verified certificate path, you will need to submit a copy of your Required Code Assignment solution into edX using a Assignment Page in the final section of each Module. These Module Code Assignments make up 20% of the Final Grade.

**Module 5**

Module 5 contains the Final Coding Assignment which is required for the verified certificate and is 30% of the Final Grade.
Course Outline

This course includes the following four modules.

- Module 1 | Sequence indexes (with Strings)
Objectives

    -    Accessing String Characters with index
    -    Accessing substrings with index slicing
    -    Iterating through Characters of a String


- Module 2 | Sequence Manipulation (with Lists)
Objectives

    -    Create Lists
    -    Access Lists
    -    Append to Lists
    -    Insert into Lists
    -    Delete Lists and List Items


- Module 3 | Sequence Iteration
Objectives

    -    Create numeric iteration using range
    -    Extend, Sort and Reverse Lists
    -    Convert between Strings and Lists
    -    Split and Join Lists


- Module 4 | Working with Files
Objectives

    -    Import, Open and Read Files
    -    Read Files a line at a time
    -    Write to Files
    -    Navigate through Files
